Import libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load data
This library has its own datasets


```python
tips=sns.load_dataset('tips')
tips.head()
```

Scatter plot of Total Bill vs Tip

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatter plot of Total Bill vs Tip')
plt.show()
```

Histogram of Total Bills

```python
sns.histplot(tips['total_bill'],bins=20,kde=True)
plt.title('Histogram of Total Bills')
plt.show()
```

Boxplot of Tips by Day

```python
sns.boxplot(x='day',y='tip',data=tips)
plt.title('Boxplot of Tips by Day')
plt.show()
```

Pairplot of Tips Dataset

```python
sns.pairplot(tips)
plt.show()
```

Customized Scatter plot of Total Bill Vs Tip

```python
sns.set_style("whitegrid")
sns.scatterplot(x='total_bill',y='tip', data=tips,color='red')
plt.title('Customized Scatter plot of Total Bill Vs Tip')
plt.show()
```

```python
tips['color']=tips['tip'].apply(lambda x:'red'if x>5 else 'green')
sns.set_style("whitegrid")
sns.scatterplot(x='total_bill',y='tip', data=tips,hue='color',palette={'red':'red','green':'green'})
plt.title('Customized Scatter plot of Total Bill Vs Tip')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='total_bill',y='tip', data=tips,color='green'.label)
plt.title('Customized Scatter plot of Total Bill Vs Tip')
plt.show()
```

Violin Plot of Total Bills by Day and Time

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='day',y='tip', hue='time',data=tips,split=True)
plt.title('Violin Plot of Total Bills by Day and Time')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bills')
plt.show()
```

Average Tip Amount by Day and Sex

```python
plt.figure(figsize=(10,6))
sns.barplot(x='day',y='tip', hue='sex',data=tips,palette='pastel',errorbar=None)
plt.title('Average Tip Amount by Day and Sex')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip')
plt.show()
```

Count of diners by Day and Smoking Preference

```python
plt.figure(figsize=(10,6))
sns.countplot(x='day', hue='smoker',data=tips,palette='coolwarm')
plt.title('Count of diners by Day and Smoking Preference')
plt.xlabel('Day of the Week')
plt.ylabel('Count of Diners')
plt.legend(title='Smoker')
plt.show()
```
