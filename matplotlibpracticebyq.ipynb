## Scatter

Load Data

```python
df=pd.read_csv('tdata.csv')
df
```

```python
x=df['Age']
```

```python
y=df['Fare']
```

```python
plt.title("Titanic Data")
plt.xlabel("Age",)
plt.ylabel("Fare")
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,c='purple',label='Age')
plt.scatter(x,y,c='green',label='Fare',alpha=0.5)

plt.title("Titanic Data")
plt.xlabel("Age",)
plt.ylabel("Fare")

plt.legend(['Age(purple)','Fare(green)'])
plt.show()
```

```python
plt.scatter(x,y,edgecolor='purple',facecolor='green')
plt.title("Scatter plt of Age vs fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
x=df['Age']
y=df['Fare']
colors=['red'if Age>30 else 'green' for Age in x]
plt.scatter(x,y,c=colors)
plt.show()
```

```python

colors = [
    'green' if 11 < age <= 20 
    else 'red' if 0 < age <= 10 
    else 'yellow' if 20 < age <= 30 
    else 'brown' if 30 < age <= 40 
    else 'orange' if 40 < age <= 50 
    else 'purple' if 50 < age <= 60 
    else 'black' if 60 < age <= 70 
    else 'blue'
    for age in x
]

plt.scatter(x,y, color=colors,marker='+')
plt.title('Age & Fare Scatter plot')
plt.xlabel('Age', color= 'Yellow')
plt.ylabel('Fare', color = 'Red')
plt.show()
```

```python
import random
x=df['Age']
y=df['Fare']
colors=[f'#{random.randint(0, 0xFFFFFF):06x}' for _ in range(len(x))]
plt.scatter(x,y, color=colors,marker='+')
plt.title('Age & Fare Scatter plot')
plt.xlabel('Age', color= 'Yellow')
plt.ylabel('Fare', color = 'Red')
plt.show()
```
