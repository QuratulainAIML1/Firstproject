```python
import tkinter as tk
from tkinter import messagebox
import random

# Define a dictionary for predefined questions and answers
responses = {
    "hi": ["Hello!", "Hi there!", "Greetings!"],
    "how are you": ["I'm doing well, thank you!", "I'm great, how about you?", "I'm fine, thanks for asking!"],
    "bye": ["Goodbye!", "See you later!", "Bye! Take care!"],
    "default": ["Sorry, I didn't understand that.", "Could you rephrase that?", "I'm not sure how to respond to that."],
    "your name":["I'm your friendly python  ChatBot."],
    "help":["Sure,I can assist you.Ask me anything"],
    "your developer" : [" I was developed by an awesome developer Quratulain."],
    "exit":["Goodbye! Have a great day."]
}

# Function to get a random response from the list of responses for a specific user input
def get_response(user_input):
    # Convert user input to lowercase for case-insensitive matching
    user_input = user_input.lower()
    
    # Check if the user input exists in the responses dictionary
    if user_input in responses:
        return random.choice(responses[user_input])
    else:
        return random.choice(responses["default"])  

# Function to handle sending a message from the user to the chatbot
def send_message():
    user_input = entry.get()  # Get user input from the entry field
    
    if user_input.lower() == "bye":
        response = "Goodbye! Have a great day!"
        text_area.config(state=tk.NORMAL)
        text_area.insert(tk.END, "You: " + user_input + "\n")
        text_area.insert(tk.END, "Chatbot: " + response + "\n")
        text_area.config(state=tk.DISABLED)
        root.quit()  # Close the application when 'bye' is typed
    else:
        response = get_response(user_input)
        text_area.config(state=tk.NORMAL)
        text_area.insert(tk.END, "You: " + user_input + "\n")
        text_area.insert(tk.END, "Chatbot: " + response + "\n")
        text_area.config(state=tk.DISABLED)
        entry.delete(0, tk.END)  # Clear the input field

# Create the main window
root = tk.Tk()
root.title("Chatbot")

# Create the text area to display the conversation
text_area = tk.Text(root, height=20, width=50, wrap=tk.WORD, bg="lightgreen",fg="black",state=tk.DISABLED)
text_area.grid(row=0, column=0, columnspan=2)

# Create the entry field for user input
entry = tk.Entry(root, width=40)
entry.grid(row=1, column=0)

# Create the send button
send_button = tk.Button(root, text="Send", width=10,bg="lightblue",fg="black", command=send_message)
send_button.grid(row=1, column=1)


# Start the main loop of the Tkinter application
root.mainloop()
```
